# General Approach

## Ausgangslage

Der vorliegende Datensatz entstand durch Webscrapping der Angebote verschiedener Händler/ Plattformen für Automobile entstanden ist. Hierbei kommt es zu Dubletten, insbesondere wenn das gleiche Fahrzeug auf mehreren Plattformen angeboten wird. 

Ziel ist es nun, diese Dubletten zu identifizieren. Die Herausforderung ist hier, dass wir zunächst keine gelabelten Daten haben. Desweiteren, dass wir nicht wie im klassischen Machine Learning Klassen lernen können. Das heisst also, Klassifikationsalgorithmen/supervised learning scheiden zunächst aus. 

Es muss also über "Ähnlichkeit" der Einträge nach Dubletten gesucht werden. Da die Daten aus unterschiedlichen Quellen stammen, müssen wir davon ausgehen, dass es systematische Unterschiede in den Daten gibt, die sich aus den eigenarten der unterschiedlichen Plattformen ergeben. Dies kann während des Scrappings versucht werden zu minimieren, aber da wir können nicht davon ausgehen, dass dies zu 100% erfolgreich ist.

Wenn es echte Aussreisser sind, dann verringern wir die chance der Detection, da gleiche aber ungewöhnliche Werte ein starkes Signal sind. # Algorithms

Wir setzen nun Fuzzy Matching ein, beim Fuzzy Matching versuchen wir Matches zwischen nicht identischen, aber ähnlichen Entitäten zu erziehlen. Eine Typische Anwendung sind spell checker/ Autokorrekturen. Wir verwenden ein Fuzzy Matching, da gleiche Autos auf unterschiedlichen Plattformen zu unterschiedlichen Einträgen führen können. 

Eine Herausforderung ist, dass wir im Prinzip jedes Element gegen jedes andere Element prüfen müssten. Ein workaround ist das kreieren von Blocks. Die Blocks können zum Beispiel durch exaktes Matching sehr aussagekräftiger Features erstellt werden. Hier bieten sich zum Beispiel Modell und Make an. Wir bilden also zum Beispiel Blöcke aus Einträgen mit gleichem Hersteller, Modell, ... und machen die Paarweise vergleiche nur innerhalb der Blöcke. 



```python
import pandas as pd
from itertools import combinations
from rapidfuzz import fuzz


def clean_text(value: object) -> str:
    """
    Clean a text value for easier comparison.

    This function:
    - turns missing values into an empty string
    - converts text to uppercase
    - removes leading and trailing spaces
    - replaces multiple spaces with a single space

    Parameters
    ----------
    value : object
        The input value from the dataframe cell.

    Returns
    -------
    str
        The cleaned text.
    """
    if pd.isna(value):
        return ""
    value = str(value).upper().strip()
    value = " ".join(value.split())
    return value


def close_enough(a: float | int | None, b: float | int | None, tolerance: float) -> bool:
    """
    Check whether two numeric values are close enough to each other.

    This is used for fuzzy matching of numeric columns like mileage,
    power, or price.

    Parameters
    ----------
    a : float | int | None
        First numeric value.
    b : float | int | None
        Second numeric value.
    tolerance : float
        Maximum allowed absolute difference.

    Returns
    -------
    bool
        True if both values exist and their absolute difference is
        smaller than or equal to the tolerance. Otherwise False.
    """
    if pd.isna(a) or pd.isna(b):
        return False
    return abs(a - b) <= tolerance


def find_duplicate_groups(
    df: pd.DataFrame,
    mileage_tolerance: float = 5000,
    power_tolerance: float = 10,
    price_tolerance: float = 2000,
    version_similarity_threshold: int = 85,
    title_similarity_threshold: int = 85,
    score_threshold: int = 4,
    max_block_size: int = 150,
) -> list[list[object]]:
    """
    Find likely duplicate car listings and return them as groups of object_ids.

    The procedure is:
    1. Clean important text columns.
    2. Match rows with the same VIN exactly.
    3. For rows without an exact VIN match, compare only rows inside
       blocks with the same make, model, registration year, and fuel.
    4. Give points for similar text and close numeric values.
    5. If enough points are collected, treat the rows as duplicates.
    6. Merge connected duplicate pairs into groups.

    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe containing car listing data. Expected columns include:
        - object_id
        - make
        - model
        - version
        - vehicle identification number
        - first_registration_year
        - first_registration_month
        - gearing
        - fuel
        - mileage
        - power
        - last_price
        - advert_title

    mileage_tolerance : float, default=5000
        Maximum allowed difference in mileage for two cars to still count
        as similar on this feature.

    power_tolerance : float, default=10
        Maximum allowed difference in power for two cars to still count
        as similar on this feature.

    price_tolerance : float, default=2000
        Maximum allowed difference in last_price for two cars to still count
        as similar on this feature.

    version_similarity_threshold : int, default=85
        Minimum fuzzy token similarity score for the version column.

    title_similarity_threshold : int, default=85
        Minimum fuzzy token similarity score for the advert_title column.

    score_threshold : int, default=4
        Minimum number of similarity points required to classify a pair
        as duplicates.

    max_block_size : int, default=150
        Maximum size of a comparison block. Larger blocks are skipped to
        keep runtime manageable.

    Returns
    -------
    list[list[object]]
        A list of duplicate groups. Each inner list contains the object_id
        values of rows that are likely duplicates.

    Example
    -------
    [[101, 205], [88, 91, 144]]
    """
    df = df.copy()

    # 1) clean important text columns
    text_columns = ["make", "model", "version", "fuel", "gearing", "advert_title"]

    for col in text_columns:
        if col in df.columns:
            df[col] = df[col].apply(clean_text)

    # 2) clean VIN if present
    if "vehicle identification number" in df.columns:
        df["vehicle identification number"] = (
            df["vehicle identification number"]
            .fillna("")
            .astype(str)
            .str.upper()
            .str.replace(r"[^A-Z0-9]", "", regex=True)
        )

    # 3) store matching pairs here
    matches: list[tuple[object, object]] = []

    # 4) exact VIN duplicates
    if "vehicle identification number" in df.columns:
        vin_groups = df[df["vehicle identification number"] != ""].groupby("vehicle identification number")

        for _, group in vin_groups:
            ids = list(group["object_id"])
            if len(ids) > 1:
                for pair in combinations(ids, 2):
                    matches.append(pair)

    # 5) fuzzy matching inside blocks
    grouped = df.groupby(["make", "model", "first_registration_year", "fuel", "gearing"])

    for _, group in grouped:
        if len(group) < 2:
            continue

        # skip very large groups
        if len(group) > max_block_size:
            continue

        rows = group.to_dict("records")

        for car1, car2 in combinations(rows, 2):
            score = 0

            # text similarity
            if fuzz.token_set_ratio(car1["version"], car2["version"]) >= version_similarity_threshold:
                score += 1

            if fuzz.token_set_ratio(car1["advert_title"], car2["advert_title"]) >= title_similarity_threshold:
                score += 1

            # numeric similarity
            if close_enough(car1["mileage"], car2["mileage"], mileage_tolerance):
                score += 1

            if close_enough(car1["power"], car2["power"], power_tolerance):
                score += 1

            if close_enough(car1["last_price"], car2["last_price"], price_tolerance):
                score += 1

            # exact matches on some fields
            if car1["gearing"] == car2["gearing"] and car1["gearing"] != "":
                score += 1

            if car1["first_registration_month"] == car2["first_registration_month"]:
                score += 1

            # enough agreement -> likely duplicate
            if score >= score_threshold:
                matches.append((car1["object_id"], car2["object_id"]))

    # 6) merge matching pairs into groups
    groups: list[set[object]] = []

    for id1, id2 in matches:
        placed = False

        for group in groups:
            if id1 in group or id2 in group:
                group.add(id1)
                group.add(id2)
                placed = True
                break

        if not placed:
            groups.append(set([id1, id2]))

    # 7) merge overlapping groups
    changed = True
    while changed:
        changed = False
        new_groups: list[set[object]] = []

        while groups:
            first = groups.pop(0)
            merged = False

            for i in range(len(groups)):
                if first.intersection(groups[i]):
                    groups[i] = first.union(groups[i])
                    merged = True
                    changed = True
                    break

            if not merged:
                new_groups.append(first)

        groups = new_groups

    # 8) convert to sorted list of lists
    result: list[list[object]] = []
    for group in groups:
        if len(group) > 1:
            result.append(sorted(list(group)))

    return result
```

```python
duplicate_groups = find_duplicate_groups(
    df,
    mileage_tolerance = 1000,
    power_tolerance = 5,
    price_tolerance = 500,
    version_similarity_threshold = 95,
    title_similarity_threshold = 95,
    score_threshold = 6
)
print(duplicate_groups[:10])
```

# Cosine Similarity



```python
import pandas as pd
from itertools import combinations
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


def clean_text(value: object) -> str:
    """
    Convert text to uppercase, strip spaces, and replace missing values with "".
    """
    if pd.isna(value):
        return ""
    value = str(value).upper().strip()
    value = " ".join(value.split())
    return value


def close_enough(a: float | int | None, b: float | int | None, tolerance: float) -> bool:
    """
    Return True if two numeric values differ by at most tolerance.
    """
    if pd.isna(a) or pd.isna(b):
        return False
    return abs(a - b) <= tolerance


def find_duplicate_pairs_cosine(
    df: pd.DataFrame,
    cosine_threshold: float = 0.92,
    mileage_bin_size: int = 1000,
    power_bin_size: int = 5,
    price_tolerance: int = 1000,
    displacement_tolerance: int = 100,
    max_block_size: int = 30,
) -> list[tuple[object, object]]:
    """
    Find likely duplicate car pairs using:
    1. exact blocking on strong structured fields
    2. cosine similarity on version + title
    3. hard numeric checks

    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe with car listing data.

    cosine_threshold : float, default=0.92
        Minimum cosine similarity required for a pair to be considered.

    mileage_bin_size : int, default=1000
        Bin size for mileage-based blocking.

    power_bin_size : int, default=5
        Bin size for power-based blocking.

    price_tolerance : int, default=1000
        Maximum allowed difference in last_price.

    displacement_tolerance : int, default=100
        Maximum allowed difference in displacement.

    max_block_size : int, default=30
        Maximum allowed block size. Larger blocks are skipped.

    Returns
    -------
    list[tuple[object, object]]
        List of matched object_id pairs.
    """
    df = df.copy()

    text_columns = ["make", "model", "version", "fuel", "gearing", "advert_title"]
    for col in text_columns:
        if col in df.columns:
            df[col] = df[col].apply(clean_text)

    if "vehicle identification number" in df.columns:
        df["vehicle identification number"] = (
            df["vehicle identification number"]
            .fillna("")
            .astype(str)
            .str.upper()
            .str.replace(r"[^A-Z0-9]", "", regex=True)
        )

    # Tighter blocking columns
    df["mileage_block"] = (df["mileage"] / mileage_bin_size).round() * mileage_bin_size
    df["power_block"] = (df["power"] / power_bin_size).round() * power_bin_size

    # Only use the strongest text fields for cosine
    df["match_text"] = (
        df["version"].fillna("").astype(str) + " " +
        df["advert_title"].fillna("").astype(str)
    ).str.strip()

    matches: list[tuple[object, object]] = []

    # 1) Exact VIN matches
    if "vehicle identification number" in df.columns:
        vin_groups = df[df["vehicle identification number"] != ""].groupby("vehicle identification number")
        for _, group in vin_groups:
            ids = list(group["object_id"])
            if len(ids) > 1:
                for pair in combinations(ids, 2):
                    matches.append(pair)

    # 2) Strict blocking
    block_cols = [
        "make",
        "model",
        "first_registration_year",
        "first_registration_month",
        "fuel",
        "power_block",
        "mileage_block",
    ]

    grouped = df.groupby(block_cols, dropna=True)

    for _, group in grouped:
        if len(group) < 2:
            continue

        if len(group) > max_block_size:
            continue

        group = group.reset_index(drop=True)

        texts = group["match_text"].fillna("").tolist()

        # Keep only rows with non-empty text
        valid_indices = []
        for i, text in enumerate(texts):
            if text.strip() != "":
                valid_indices.append(i)

        if len(valid_indices) < 2:
            continue

        group = group.iloc[valid_indices].reset_index(drop=True)
        texts = group["match_text"].tolist()

        try:
            vectorizer = TfidfVectorizer()
            matrix = vectorizer.fit_transform(texts)
        except ValueError:
            continue

        sim_matrix = cosine_similarity(matrix)

        for i in range(len(group)):
            for j in range(i + 1, len(group)):
                if sim_matrix[i, j] < cosine_threshold:
                    continue

                car1 = group.iloc[i]
                car2 = group.iloc[j]

                # Optional extra numeric sanity checks
                if not close_enough(car1.get("last_price"), car2.get("last_price"), price_tolerance):
                    continue

                if not close_enough(car1.get("displacement"), car2.get("displacement"), displacement_tolerance):
                    continue

                matches.append((car1["object_id"], car2["object_id"]))

    return matches
```

```python
duplicate_pairs = find_duplicate_pairs_cosine(
    df,
    cosine_threshold=0.95,
    mileage_bin_size=500,
    power_bin_size=5,
    price_tolerance=500,
    displacement_tolerance=50,
    max_block_size=20,
)

print("number of pairs:", len(duplicate_pairs))
print(duplicate_pairs[:20])
```
